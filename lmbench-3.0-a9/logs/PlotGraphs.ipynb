```python
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cur_path = os.getcwd()
print(cur_path)
df = pd.read_csv("../stats.csv")
df
```

```python
df_sorted = df.sort_values(by=['Data','No.of_Parallel_Processes'])
df_sorted
```

```python
def group_by_plot(kind:str,x:str,y:str,title:str):
    '''
        This method plots the graph using Data as labels. It accepts two parameters x and y
        which are used to plot the graph.

        Params:
            kind: (str) What type of plot
            x: Column name in pandas df which is assigned x-axis
            y: Column name in pandas df which is assigned y-axis
            title: (str) Title for the plot
    '''
    fig, ax = plt.subplots()
    colors = {10.0:'red', 20.0:'green', 30.0:'blue', 40.0:'black'}
    grouped = df_sorted.groupby('Data')
    for key, group in grouped:
        group.plot(ax=ax, kind=kind, x=x, y=y, label=key, color=colors[key], title=title)

    plt.savefig(title+".png")
    plt.show()
```

```python
# Plot Number of Parallel Processes vs Time Elapsed
group_by_plot("line","No.of_Parallel_Processes","Time_Elapsed","No. Parallel Processes vs Time Elapsed")
```

```python
# Plot Number of Parallel Processes vs Bandwidth
group_by_plot("line","No.of_Parallel_Processes","Bandwidth","No. Parallel Processes vs Bandwidth")
```

```python
# Plot Number of Parallel Processes vs Time Taken
group_by_plot("line","No.of_Parallel_Processes","Time_Taken","No. Parallel Processes vs Time Taken")
```
